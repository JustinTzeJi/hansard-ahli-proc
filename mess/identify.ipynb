```python
df2 = pd.read_csv("DR - dr_1_11_id_all.csv")
```

```python
df2 = df2.drop(columns="index")
```

```python
df2
```

```python
df3 = df2[df2["id"]!=9999999][["vol", "id"]].groupby("id").min().reset_index().sort_values(["vol", "id"]).reset_index(drop=True).reset_index()[["index","id"]]
```

```python
df4 = pd.merge(df2, df3, "left", on="id")
```

```python
df5 = df4[["tahun_lahir","index"]].dropna().drop_duplicates()
```

```python
double_index = df5["index"].value_counts()[(df5["index"].value_counts()>1)].index
```

```python
df5[df5["index"]==473]
```

```python
df4[df4["index"] == 1130]
```

```python
double_index_fix = df5[df5["index"].isin(double_index)].groupby("index").min().reset_index()
```

```python
double_index_fix
```

```python
single_index = df5[~df5["index"].isin(double_index)]
```

```python
tahun_lahir_map = pd.concat([single_index,double_index_fix]).sort_values("index").reset_index(drop=True)
```

```python
tahun_lahir_map
```

```python
df4.drop(columns="tahun_lahir").merge(tahun_lahir_map, on="index", how="left").to_csv("master_table.csv", index=False)
```
